## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
z = [6,7,8,9,10]
```

```python
plt.plot(x,y, color="red", linestyle="--", marker="o")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Basic Line Plot")
plt.grid(True)
```

```python
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.plot(x,y,color="red",marker="o")
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
plt.title("Red")

plt.subplot(2,2,2)
plt.plot(y,x,color="green",marker="o")
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
plt.title("Green")

plt.subplot(2,2,3)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.plot(x,z,color="blue",marker="o")
plt.title("Blue")

plt.subplot(2,2,4)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.plot(z,x,color="orange",marker="o")
plt.title("Ornage")
```

## Barplot


```python
categories = ['A','B','C','D','E']
values = [5,7,3,8,6]
plt.bar(categories, values)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Plot")
```

## Histogram

```python
data = [1,2,2,3,3,4,5,6,6,6,6,7]

plt.hist(data, bins=7,color="orange",edgecolor="black")
```

## Scatterplot

```python
x = [1,2,3,4,5,6]
y = [8,9,4,9,10,7]

plt.scatter(x,y)
```

## Pie Chart

```python
labels = ['A','B','C','D']   # match sizes length
sizes = [30,20,40,10]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.1,0,0,0)  # explode first slice

plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True)
plt.axis('equal')  # ensures pie is drawn as a circle
plt.show()
```
